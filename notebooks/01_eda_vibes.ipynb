```python
import pandas as pd

df = pd.read_csv("../data/spotify_tracks.csv")
df.head()
df.info()
df.describe()[["danceability","energy","valence","acousticness","instrumentalness","liveness","speechiness","tempo"]]
df["track_genre"].value_counts().head(20)
```
